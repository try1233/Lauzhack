```python
import pandas as pd

# Specify the file path of the dataset
file_path = 'data/Amazon_Unlocked_Mobile.csv'

# Read the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
pd.set_option('display.max_colwidth', None)
```

```python
df
```

```python
df['Combined Info'] = df['Brand Name'] + ' ' + df['Product Name'] + ' Price:' + df['Price'].astype(str) + ' Review: ' + df['Reviews']
```

```python
import matplotlib.pyplot as plt

plot = df['Product Name'].value_counts().head(200).plot(kind='bar')
plot.set_xticklabels(['Item {}'.format(i+1) for i in range(200)])
plot.set_xticks(range(0, 200, 25))

plt.show()
```
